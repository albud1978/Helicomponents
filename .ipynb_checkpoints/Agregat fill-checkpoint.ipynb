```python
from clickhouse_driver import Client
import os

# Получение параметров подключения из переменных окружения
clickhouse_host = os.getenv('CLICKHOUSE_HOST', '10.95.19.132')
clickhouse_user = os.getenv('CLICKHOUSE_USER', 'default')
clickhouse_password = os.getenv('CLICKHOUSE_PASSWORD', 'quie1ahpoo5Su0wohpaedae8keeph6bi')
database_name = os.getenv('CLICKHOUSE_DB', 'default')

# Подключение к ClickHouse
client = Client(
    host=clickhouse_host,
    user=clickhouse_user,
    password=clickhouse_password,
    database=database_name
)

try:
    # Выполнение вычислений и запись в таблицу Agregat
    get_first_row_values_query = """
        SELECT ComponentNumber, NR_Mi8T, MRR_Mi8T, NR_Mi17, MRR_Mi17, PurchasePrice, RepairPrice
        FROM Agregat
        LIMIT 1
    """

    first_row_values = client.execute(get_first_row_values_query)

    if first_row_values:
        # Распаковываем значения, включая ComponentNumber
        ComponentNumber, NR_Mi8T, MRR_Mi8T, NR_Mi17, MRR_Mi17, PurchasePrice, RepairPrice = first_row_values[0]

        # Логирование извлечённых значений
        print(f"ComponentNumber: {ComponentNumber}, NR_Mi8T: {NR_Mi8T}, MRR_Mi8T: {MRR_Mi8T}, NR_Mi17: {NR_Mi17}, MRR_Mi17: {MRR_Mi17}, PurchasePrice: {PurchasePrice}, RepairPrice: {RepairPrice}")

        # Проверка на нулевые значения для предотвращения деления на ноль
        if MRR_Mi8T != 0 and MRR_Mi17 != 0:
            # Вычисление BR_Mi8T и BR_Mi17
            BR_Mi8T = NR_Mi8T - (RepairPrice / ((PurchasePrice - RepairPrice) / NR_Mi8T + RepairPrice / MRR_Mi8T))
            BR_Mi17 = NR_Mi17 - (RepairPrice / ((PurchasePrice - RepairPrice) / NR_Mi17 + RepairPrice / MRR_Mi17))

            # Округление значений до целых чисел
            BR_Mi8T = round(BR_Mi8T)
            BR_Mi17 = round(BR_Mi17)

            # Логирование вычисленных и округленных значений
            print(f"Вычисленное значение BR_Mi8T (до округления): {BR_Mi8T}")
            print(f"Вычисленное значение BR_Mi17 (до округления): {BR_Mi17}")

            # Обновление таблицы Agregat значениями BR_Mi8T и BR_Mi17 для первой строки
            update_query = """
                ALTER TABLE Agregat 
                UPDATE BR_Mi8T = %(BR_Mi8T)s, BR_Mi17 = %(BR_Mi17)s 
                WHERE ComponentNumber = %(ComponentNumber)s
            """
            params = {
                'BR_Mi8T': BR_Mi8T,
                'BR_Mi17': BR_Mi17,
                'ComponentNumber': ComponentNumber
            }
            client.execute(update_query, params)
            print("Поля BR_Mi8T и BR_Mi17 успешно обновлены в таблице Agregat.")
        else:
            print("Невозможно выполнить деление на ноль. Проверьте значения MRR_Mi8T и MRR_Mi17.")
    else:
        print("Нет значений для обновления в таблице Agregat.")
except Exception as e:
    print(f"Произошла ошибка: {e}")
```
