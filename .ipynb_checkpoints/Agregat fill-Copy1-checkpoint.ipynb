```python
from clickhouse_driver import Client
import os

# Получение параметров подключения из переменных окружения
clickhouse_host = os.getenv('CLICKHOUSE_HOST', '10.95.19.132')
clickhouse_user = os.getenv('CLICKHOUSE_USER', 'default')
clickhouse_password = os.getenv('CLICKHOUSE_PASSWORD', 'quie1ahpoo5Su0wohpaedae8keeph6bi')
database_name = os.getenv('CLICKHOUSE_DB', 'default')

# Подключение к ClickHouse
client = Client(
    host=clickhouse_host,
    user=clickhouse_user,
    password=clickhouse_password,
    database=database_name
)

try:
    # Выполнение вычислений и запись в таблицу Agregat
    get_first_row_values_query = """
        SELECT ComponentNumber, NR_Mi8T, MRR_Mi8T, NR_Mi17, MRR_Mi17, PurchasePrice, RepairPrice
        FROM Agregat
        LIMIT 1
    """

    first_row_values = client.execute(get_first_row_values_query)

    if first_row_values:
        # Распаковываем значения, включая ComponentNumber
        ComponentNumber, NR_Mi8T, MRR_Mi8T, NR_Mi17, MRR_Mi17, PurchasePrice, RepairPrice = first_row_values[0]

        # Логирование извлечённых значений
        print(f"ComponentNumber: {ComponentNumber}, NR_Mi8T: {NR_Mi8T}, MRR_Mi8T: {MRR_Mi8T}, NR_Mi17: {NR_Mi17}, MRR_Mi17: {MRR_Mi17}, PurchasePrice: {PurchasePrice}, RepairPrice: {RepairPrice}")

        # Проверка на нулевые значения для предотвращения деления на ноль
        if MRR_Mi8T != 0 and MRR_Mi17 != 0:
            # Вычисление BR_Mi8T и BR_Mi17
            BR_Mi8T = NR_Mi8T - (RepairPrice / ((PurchasePrice - RepairPrice) / NR_Mi8T + RepairPrice / MRR_Mi8T))
            BR_Mi17 = NR_Mi17 - (RepairPrice / ((PurchasePrice - RepairPrice) / NR_Mi17 + RepairPrice / MRR_Mi17))

            # Округление значений до целых чисел
            BR_Mi8T = round(BR_Mi8T)
            BR_Mi17 = round(BR_Mi17)

            # Логирование вычисленных и округленных значений
            print(f"Вычисленное значение BR_Mi8T (до округления): {BR_Mi8T}")
            print(f"Вычисленное значение BR_Mi17 (до округления): {BR_Mi17}")

            # Обновление таблицы Agregat значениями BR_Mi8T и BR_Mi17 для первой строки
            update_query = """
                ALTER TABLE Agregat 
                UPDATE BR_Mi8T = %(BR_Mi8T)s, BR_Mi17 = %(BR_Mi17)s 
                WHERE ComponentNumber = %(ComponentNumber)s
            """
            params = {
                'BR_Mi8T': BR_Mi8T,
                'BR_Mi17': BR_Mi17,
                'ComponentNumber': ComponentNumber
            }
            client.execute(update_query, params)
            print("Поля BR_Mi8T и BR_Mi17 успешно обновлены в таблице Agregat.")
        else:
            print("Невозможно выполнить деление на ноль. Проверьте значения MRR_Mi8T и MRR_Mi17.")
    else:
        print("Нет значений для обновления в таблице Agregat.")
except Exception as e:
    print(f"Произошла ошибка: {e}")
```

```python
import numpy as np
import pandas as pd
import logging
from clickhouse_driver import Client

# Настройки подключения к ClickHouse
clickhouse_host = '10.95.19.132'
clickhouse_port = 9000
clickhouse_user = 'default'
clickhouse_password = 'quie1ahpoo5Su0wohpaedae8keeph6bi'
database_name = 'default'
table_name = 'Vygruzka'
agregat_table = 'Agregat'

# Настройка логирования
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Функция для создания нового поля BR
def calculate_br(client):
    logging.info("Начало расчета поля BR")
    # Загрузка данных из таблицы Vygruzka
    vygruzka_query = f"SELECT ll, oh FROM {database_name}.{table_name}"
    try:
        vygruzka_data = client.execute(vygruzka_query)
    except Exception as e:
        logging.error(f"Ошибка при загрузке данных из таблицы Vygruzka: {e}")
        return

    vygruzka_df = pd.DataFrame(vygruzka_data, columns=['ll', 'oh'])

    # Загрузка фиксированных данных из таблицы Agregat
    agregat_query = f"SELECT RepairPrice, PurchasePrice FROM {database_name}.{agregat_table} LIMIT 1"
    try:
        agregat_data = client.execute(agregat_query)
    except Exception as e:
        logging.error(f"Ошибка при загрузке данных из таблицы Agregat: {e}")
        return

    agregat_df = pd.DataFrame(agregat_data, columns=['RepairPrice', 'PurchasePrice'])

    if agregat_df.empty:
        logging.error("Таблица Agregat не содержит данных")
        return

    repair_price = agregat_df.at[0, 'RepairPrice']
    purchase_price = agregat_df.at[0, 'PurchasePrice']

    # Вычисление поля BR по заданной формуле для каждой записи в Vygruzka
    def calculate_br_row(row):
        ll = row['ll']
        oh = row['oh']

        if pd.isna(oh) or oh == 0:
            return 0
        else:
            try:
                return round(ll - (repair_price / ((purchase_price - repair_price) / ll + repair_price / oh)))
            except ZeroDivisionError:
                return 0

    vygruzka_df['BR'] = vygruzka_df.apply(calculate_br_row, axis=1)

    # Запись данных обратно в таблицу ClickHouse
    try:
        client.execute(f"ALTER TABLE {database_name}.{table_name} ADD COLUMN IF NOT EXISTS BR Float64")
    except Exception as e:
        logging.error(f"Ошибка при создании поля BR: {e}")
        return

    # Обновление значений в таблице Vygruzka
    for _, row in vygruzka_df.iterrows():
        try:
            update_query = f"ALTER TABLE {database_name}.{table_name} UPDATE BR = {row['BR']} WHERE ll = {row['ll']} AND oh = {row['oh']}"
            client.execute(update_query)
        except Exception as e:
            logging.error(f"Поле BR не создано для строки с ll={row['ll']} и oh={row['oh']}: {e}")

    logging.info("Расчет поля BR завершен успешно")

# Подключение к ClickHouse и вычисление поля BR
client = Client(host=clickhouse_host, port=clickhouse_port, user=clickhouse_user, password=clickhouse_password)
calculate_br(client)
```
